```python
import pandas as pd

# Load data
df = pd.read_csv('h.pylori.csv')

# Clean up data
df['test result '] = df['test result '].replace({2.0: 1.0})
df = df.drop(columns=['treatment', 'treatment result'])
df = df.iloc[1:, :]
df = df.dropna()

# Encode job categories
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['job_encoded'] = label_encoder.fit_transform(df['job'])
df = df.drop(columns=['job','case no'])
print(df.columns)
print('\n\n\n\n')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Prepare the cleaned data
X_clean = df.drop(columns=['test result '], axis=1)
y_clean = df['test result ']
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.3, random_state=42)

# Initialize the SVC model
svc_model = SVC()

# Fit the SVC model with the cleaned data
svc_model.fit(X_train, y_train)

# Evaluate the SVC model
y_pred = svc_model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

# Initialize GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)

# Fit the grid search to the data
grid.fit(X_train, y_train)

# Print the best parameters and the best score
print(f"Best parameters found: {grid.best_params_}")
print(f"Best cross-validation score: {grid.best_score_}")

# Evaluate the best model
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
Q1 = df['months'].quantile(0.25)
Q3 = df['months'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['months'] >= lower_bound) & (df['months'] <= upper_bound)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Prepare the cleaned data
X_clean = df.drop(columns=['test result '], axis=1)
y_clean = df['test result ']
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.3, random_state=42)

# Initialize the SVC model
svc_model = SVC()

# Fit the SVC model with the cleaned data
svc_model.fit(X_train, y_train)

# Evaluate the SVC model
y_pred = svc_model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```
