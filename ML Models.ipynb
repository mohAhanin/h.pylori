```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
# Load data
df = pd.read_csv('h.pylori.csv')

# Display info
df.info()

# Replace values in 'test result '
df['test result '] = df['test result '].replace({2.0: 1.0})
print(df['test result '].value_counts())
```

```python
# Drop unnecessary columns
df = df.drop(columns=['treatment', 'treatment result', 'malignant stomach'])
df
```

```python
# Remove first row if needed
df = df.iloc[1:, :]
df
```

```python
# Display unique job counts
print(df['job'].value_counts())
```

```python
# Encode job categories
label_encoder = LabelEncoder()
df['job_encoded'] = label_encoder.fit_transform(df['job'])
df
```

```python
# Print mapping
mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(mapping)
```

```python
print(df[['job', 'job_encoded']].drop_duplicates())
```

```python
df = df.drop(columns=['job'])
df
```

```python
# Check for negative values in the DataFrame
negative_values = (df < 0).sum().sum()
print(f"Total negative values in DataFrame: {negative_values}")
```

```python
# Display rows with negative values if any (should be none)
negative_rows = df[df < 0].dropna(how='all')
print(negative_rows)
```

```python
# Check for NaN values and handle them
print(df.isnull().sum())
df = df.dropna()
df
```

```python
df.info()
```

```python
# Define target and features
y = df['test result ']
X = df.drop(columns=['test result '], axis=1)
y
```

```python
X
```

```python
# Check feature columns
print(X.columns)
X
```

```python
# Correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Display value counts for each column
for col in df.columns:
    print(df[col].value_counts())

```

```python
# Drop 'case no' column if needed
df = df.drop(columns=['case no'])
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Initialize LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=False, custom_metric=None)
```

```python
# Fit and evaluate models
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```
