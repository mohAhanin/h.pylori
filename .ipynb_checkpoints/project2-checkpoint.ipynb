```python
import pandas as pd

data = pd.read_csv('C:/Users/IDEH/Downloads/Telegram Desktop/dataset/h.pylori.csv')

data = data.iloc[1:, :]  

data['test result '] = data['test result '].replace({2.0: 1.0})

data['epigastric pain'] = data['epigastric pain'].fillna(data['epigastric pain'].mode()[0])
data['fullness'] = data['fullness'].fillna(data['fullness'].mode()[0])
data['reflux'] = data['reflux'].fillna(data['reflux'].mode()[0])
data['smoker'] = data['smoker'].fillna(data['smoker'].mode()[0])
data['Revisit'] = data['Revisit'].fillna(data['Revisit'].mode()[0])

data['test result '] = data['test result '].fillna(0)

print("Missing values:", data.isnull().sum())

positive_ubt = data[data['test result '] == 1]

# Treatment analysis for triple therapy
# Check how many patients were treated with '3' or '34' (representing triple therapy treatments)
triple_therapy = positive_ubt[(positive_ubt['treatment'] == str(3)) | (positive_ubt['treatment'] == str(34))]

# Analyze the treatment results (assumed to be in the 'treatment result' column)
treatment_results = triple_therapy['treatment result'].value_counts()

# Calculate the clinical improvement percentage
total_triple_therapy = len(triple_therapy)
improvement_count = treatment_results.get(1, 0)  # Assuming '1' represents clinical improvement
improvement_percentage = (improvement_count / total_triple_therapy) * 100

# Output the results
print(f"Total patients treated with triple therapy: {total_triple_therapy}")
print(f"Patients with clinical improvement: {improvement_count}")
print(f"Clinical improvement percentage: {improvement_percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/IDEH/Downloads/Telegram Desktop/dataset/h.pylori.csv')

data = data.iloc[1:, :]  
data['test result '] = data['test result '].replace({2.0: 1.0})  
data['test result '] = data['test result '].fillna(0) 


positive_ubt = data[data['test result '] == 1]


triple_therapy = positive_ubt[positive_ubt['treatment'] == str(3)]

female_triple_therapy = triple_therapy[triple_therapy['sex'] == 0]  # 0 
male_triple_therapy = triple_therapy[triple_therapy['sex'] == 1]    # 1 

female_treatment_results = female_triple_therapy['treatment result'].value_counts()
male_treatment_results = male_triple_therapy['treatment result'].value_counts()

total_female_patients = len(female_triple_therapy)
if total_female_patients > 0:
    female_improvement_count = female_treatment_results.get(1, 0)  # ۱ نشان‌دهنده بهبودی
    female_improvement_percentage = (female_improvement_count / total_female_patients) * 100
else:
    female_improvement_percentage = 0

total_male_patients = len(male_triple_therapy)
if total_male_patients > 0:
    male_improvement_count = male_treatment_results.get(1, 0)  # ۱ نشان‌دهنده بهبودی
    male_improvement_percentage = (male_improvement_count / total_male_patients) * 100
else:
    male_improvement_percentage = 0

improvement_df = pd.DataFrame({
    'Gender': ['Female', 'Male'],
    'Total Patients': [total_female_patients, total_male_patients],
    'Improvement Count': [female_improvement_count, male_improvement_count],
    'Improvement Percentage': [female_improvement_percentage, male_improvement_percentage]
})

print(improvement_df)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].bar(['Female', 'Male'], [female_improvement_percentage, male_improvement_percentage], color=['pink', 'blue'])
ax[0].set_title('Clinical Improvement Percentage by Gender')
ax[0].set_ylabel('Improvement Percentage (%)')

ax[1].bar(['Female', 'Male'], [total_female_patients, total_male_patients], alpha=0.5, label='Total Patients', color=['lightpink', 'lightblue'])
ax[1].bar(['Female', 'Male'], [female_improvement_count, male_improvement_count], label='Improved Patients', color=['pink', 'blue'])
ax[1].set_title('Total Patients vs Improved Patients by Gender')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('C:/Users/IDEH/Downloads/Telegram Desktop/dataset/h.pylori.csv')

# Preprocessing data
data = data.iloc[1:, :]
data['test result '] = data['test result '].replace({2.0: 1.0})  # Replace 2 with 1 in test results
data['test result '] = data['test result '].fillna(0)  # Fill NaN values in test results

# Filter patients who received triple therapy (treatment code 3) with positive UBT test
triple_therapy = data[(data['treatment'] == str(3)) & (data['test result '] == 1)]

# Count improvement results for men and women
improvement_counts = triple_therapy.groupby('sex')['treatment result'].value_counts().unstack(fill_value=0)

# Calculate improvement percentages
improvement_percentages = (improvement_counts[1] / improvement_counts.sum(axis=1)) * 100

# Create DataFrame for results
improvement_df = pd.DataFrame({
    'Total Patients': improvement_counts.sum(axis=1),
    'Improvement Count': improvement_counts[1],
    'Improvement Percentage': improvement_percentages
}).reset_index()

# Print DataFrame
print(improvement_df)

# Plotting improvement percentages for men and women
plt.figure(figsize=(8, 6))
plt.bar(improvement_df['sex'], improvement_df['Improvement Percentage'], color=['blue', 'pink'])
plt.title('Improvement Percentage of Triple Therapy for H. pylori by Gender')
plt.xlabel('Gender')
plt.ylabel('Improvement Percentage (%)')
plt.ylim(0, 100)

# Display improvement percentage on the bars
for index, value in enumerate(improvement_df['Improvement Percentage']):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', fontsize=10)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/IDEH/Downloads/Telegram Desktop/dataset/h.pylori.csv')

data = data.iloc[1:, :]
data['test result '] = data['test result '].replace({2.0: 1.0})  # Replace 2 with 1 in test results
data['test result '] = data['test result '].fillna(0)  # Fill NaN values in test results

# Filter patients who received triple therapy (treatment code 3) with positive UBT test
triple_therapy = data[(data['treatment'] == str(3)) & (data['test result '] == 1)]

# Count improvement results for patients with and without a history of ulcers
improvement_counts = triple_therapy.groupby('stomach ulcer')['treatment result'].value_counts().unstack(fill_value=0)

# Calculate improvement percentages
improvement_percentages = (improvement_counts[1] / improvement_counts.sum(axis=1)) * 100

# Create DataFrame for results
improvement_df = pd.DataFrame({
    'Total Patients': improvement_counts.sum(axis=1),
    'Improvement Count': improvement_counts[1],
    'Improvement Percentage': improvement_percentages
}).reset_index()

# Rename columns for clarity
improvement_df['History of Ulcer'] = improvement_df['stomach ulcer'].map({0: 'No History', 1: 'History'})

# Print DataFrame
print(improvement_df[['History of Ulcer', 'Total Patients', 'Improvement Count', 'Improvement Percentage']])

# Plotting improvement percentages for patients with and without a history of ulcers
plt.figure(figsize=(8, 6))
plt.bar(improvement_df['History of Ulcer'], improvement_df['Improvement Percentage'], color=['blue', 'orange'])
plt.title('Improvement Percentage of Triple Therapy for H. pylori by History of Ulcer')
plt.xlabel('History of Ulcer')
plt.ylabel('Improvement Percentage (%)')
plt.ylim(0, 100)

# Display improvement percentage on the bars
for index, value in enumerate(improvement_df['Improvement Percentage']):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', fontsize=10)

plt.show()
```
